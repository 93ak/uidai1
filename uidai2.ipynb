```python
from datasets import load_dataset
```

```python
ds = load_dataset("an42/uidai1", data_files="bioapi1.csv")
biodf = ds["train"].to_pandas()
```

```python
bioout = (
    biodf.groupby('Pincode')
      .agg({
          'State': lambda x: set(x),
          'District': lambda x: set(x)
      })
      .to_dict('index')
)
```

```python
bioout
```

```python
bio_multi_district = sum(len(v['District']) > 1 for v in bioout.values())
bio_multi_state    = sum(len(v['State']) > 1 for v in bioout.values())

print(bio_multi_district, bio_multi_state)
```

```python
sp_low=lambda s:s.replace(' ','').lower()
```

```python
canon_keys = set(keys.values())
bio_map = {k: set() for k in canon_keys}
bio_map['others'] = set()
# bio_map={}
for raw in biodf['State']:
    # matched = False
    for k in keys:
        if k in sp_low(raw):
            # print(k,raw,keys[k])
            bio_map[keys[k]].add(raw)
            # matched = True
            break
    # if not matched:
    else:
        bio_map['others'].add(raw)
```
